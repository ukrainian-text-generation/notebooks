```python
from string import Template

LINE_TEMPLATE = Template('{"messages": [{"role": "system", "content": "Асистент є системою генерації текстів українською мовою за заданим контекстом"}, $input, $output]}')
SOURCE_DATASET_PATH = "/content/drive/MyDrive/diploma-llm/data/fine-tuning/dataset.json"
TARGET_DATASET_PATH = "/content/drive/MyDrive/diploma-llm/data/fine-tuning/gpt/dataset-cropped.jsonl"
```

```python
import json

def transform_to_jsonl(input_file, output_file):
  with open(input_file, 'r') as f:
    data = json.load(f)
  with open(output_file, 'w', encoding='utf-8') as f:
    for entry in data:
      # print(entry)
      input = {'role': 'user', 'content': entry['input'].replace('\n', ' ').replace('\r', '')}
      output = {'role': 'assistant', 'content': entry['output'].replace('\n', ' ').replace('\r', '')[:2000]}
      f.write(LINE_TEMPLATE.substitute(input=json.dumps(input, ensure_ascii=False).encode('utf-8').decode(), output=json.dumps(output, ensure_ascii=False).encode('utf-8').decode()))
      # print(json.dumps(input, ensure_ascii=False).encode('utf-8').decode())
      f.write('\n')
```

```python
transform_to_jsonl(SOURCE_DATASET_PATH, TARGET_DATASET_PATH)
```
